```python
#creating mapping from spatial to readzs cell types
spatial_to_readzs_ann = {
    'epithelial cell of proximal tubule':'Epcam    kidney proximal convoluted tubule epithelial cell',
    'glomerular capillary endothelial cell':'Pecam    kidney capillary endothelial cell',
    'kidney blood vessel cell':'CD45    plasma cell',
    'kidney collecting duct epithelial cell':'Epcam    kidney collecting duct principal cell',
    'kidney distal convoluted tubule epithelial cell':'Epcam     kidney distal convoluted tubule epithelial cell',
    'kidney loop of Henle epithelial cell':'Epcam    thick ascending tube S epithelial cell',
    'lymphocyte':'B cell',
    'macrophage':'CD45    macrophage',
    #'pericyte',
    'podocyte':'Epcam    podocyte',
}

spatial_df['ontology'] = spatial_df['ontology'].map(spatial_to_readzs_ann)
spatial_df = spatial_df.dropna()
spatial_df.head()
```

```python
#merge the spatial and readzs tables
m = spatial_df.merge(readzs_df, on=['gene','ontology'])

#only keep genes present in at least 3 ontologies
m = m.groupby('gene').filter(lambda g: g['ontology'].nunique() >= 3)

m.head()
```

```python
m.columns
```

```python
for gene,plot_g in m.groupby('gene'):
    
    r,p = stats.pearsonr(plot_g['median_z_scaled'], plot_g['med_score'])
    
    sns.scatterplot(
        x = 'median_z_scaled',
        y = 'med_score',
        hue = 'ontology',
        data = plot_g,
    )
    plt.title('{} has Pearson r={:.2f} p={:.2f}'.format(gene,r,p))
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.xlabel('Median ReadZs z-score scaled')
    plt.ylabel('Median periphery score')
    plt.show()
    plt.close()
```

# Read buildup plot for Hnf1b

```python
counts_df = pd.read_csv(
    '/scratch/groups/horence/Elisabeth/intron_zscore/Rob_data/20220331_without_mouse_id/counts/20220331_tabula_muris_kidney_liver_gene_level_no_unann_filt_chr11.txt',
    sep = '\t',
    header = None,
    names = ['cell_id','chr','pos','strand','count','sample','window'],
)

counts_df = counts_df[
    counts_df['window'].str.contains('Hnf1b') &
    counts_df['strand'].eq('+') &
    counts_df['pos'].between(83904000,83906000)
]
counts_df
```

```python
meta_df = pd.read_csv(
    '/scratch/groups/horence/Elisabeth/intron_zscore/Rob_data/20220331_without_mouse_id/zscore/20220331_tabula_muris_kidney_liver_gene_level_no_unann_filt_chr11.zscore',
    sep = '\t',
)

meta_df = meta_df.dropna(subset=['free_annotation'])
meta_df = meta_df[
    meta_df['window'].str.contains('Hnf1b') &
    meta_df['read_strand'].eq('+')
]

cell_id_to_free_annotation = dict(meta_df[['cell_id','free_annotation']].values)
cell_id_to_free_annotation
```

```python
counts_df['free_annotation'] = counts_df['cell_id'].map(cell_id_to_free_annotation)
counts_df = counts_df.dropna()
counts_df
```

```python
bin_size = 5
df = counts_df.copy()

#hardcoding Hnf1b 3' UTR
chrom = 'chr11'
min_val = 83904000
max_val = 83906000

df = df[df['pos'].between(min_val,max_val)]
df = df.groupby('free_annotation').filter(lambda g: g['count'].sum() >= 5)

bins = np.arange(min_val,max_val,bin_size)
bin_labels = pd.Categorical(bins[:-1])

df['bin'] = pd.cut(df['pos'], bins=bins)


num_onts = df['free_annotation'].nunique()
colors = sns.color_palette('viridis',num_onts)

for i,(ont,g) in enumerate(df.groupby('free_annotation')):

    red,green,blue = colors[i]
    red,green,blue = int(256*red),int(256*green),int(256*blue)

    tot_counts = g['count'].sum()

    bin_counts = g.groupby(['chr','bin'])['count'].sum().fillna(0)
    norm_bin_counts = bin_counts/bin_counts.max()
    cs_counts = bin_counts.cumsum()/tot_counts

    #Write out the cumulative table
    cs_out_path = 'tabula_sapiens_timp3_readzs/bedGraphs/Hnf1b_{}.bedGraph'.format(ont).replace(' ','_')

    with open(cs_out_path,'w') as f_out:
        f_out.write('browser position {}:{}-{}\n'.format(chrom,min_val,max_val))
        f_out.write('track type=bedGraph name="cs {}" visibility=hide maxHeightPixels=50 graphType=points color={},{},{} visibility=hide\n'.format(ont,red,green,blue))

    cs_df = cs_counts.reset_index()
    cs_df['left'] = cs_df['bin'].apply(lambda v: v.left)
    cs_df['right'] = cs_df['bin'].apply(lambda v: v.right)
    cs_df[['chr','left','right','count']].to_csv(cs_out_path, mode='a', header=None, sep=' ', index=None)

    #Write out the density table
    density_out_path = 'tabula_sapiens_timp3_readzs/bedGraphs/density_Hnf1b_{}.bedGraph'.format(ont).replace(' ','_')

    with open(density_out_path,'w') as f_out:
        f_out.write('browser position {}:{}-{}\n'.format(chrom,min_val,max_val))
        f_out.write('track type=bedGraph name="{}" visibility=full maxHeightPixels=50 graphType=points color={},{},{} visibility=hide\n'.format(ont,red,green,blue))

    density_df = norm_bin_counts.reset_index()
    density_df['left'] = density_df['bin'].apply(lambda v: v.left)
    density_df['right'] = density_df['bin'].apply(lambda v: v.right)
    density_df[['chr','left','right','count']].to_csv(density_out_path, mode='a', header=None, sep=' ', index=None)
```
