````python
from skimage.segmentation import watershed, mark_boundaries
from skimage.filters import sobel
from skimage.exposure import histogram
from scipy import ndimage as ndi
from skimage.measure import find_contours, approximate_polygon

#Following tutorial here
#https://scikit-image.org/docs/stable/user_guide/tutorial_segmentation.html
sub_img = im[8000:10000,8000:10000].copy()

hist, hist_centers = histogram(sub_img)
sns.lineplot(x = hist_centers, y = hist)
plt.title('Histogram of pixel values')
plt.show()
plt.close()

plt.imshow(sub_img)
plt.title('Before masking')
plt.show()
plt.close()

elevation_map = sobel(sub_img)

markers = np.zeros_like(sub_img)
markers[sub_img < 500] = 1
markers[sub_img > 1500] = 2


plt.imshow(markers)
plt.title('After masking')
plt.show()
plt.close()

segmentation = watershed(elevation_map, markers)
segmentation = ndi.binary_fill_holes(segmentation - 1)

plt.imshow(segmentation)
plt.title('Watershed segmented')
plt.show()
plt.close()

#THIS GIVES BACK (y,x) for some reason!!
#followed tutorial here: https://scikit-image.org/docs/dev/auto_examples/edges/plot_polygon.html#sphx-glr-auto-examples-edges-plot-polygon-py
raw_nuclei_polys = find_contours(segmentation,0)

fig = plt.plot()
plt.imshow(sub_img)
simplified_nuclei_polys = []
for raw_p in raw_nuclei_polys:
    coords = approximate_polygon(raw_p, tolerance=2.5) #simplifying the polygons saves a lot of space and looks good still
    simplified_nuclei_polys.append(coords)
    
    #plt.plot(raw_p[:,1],raw_p[:,0]) #reversed plotting than I'd expect
    plt.plot(coords[:,1],coords[:,0]) #reversed plotting than I'd expect
    
plt.show()
plt.title('Boundaries superimposed')
plt.close()
````

````python
sns.displot([p.shape[0] for p in raw_nuclei_polys])
plt.title('Distribution of number of vertices per nuclei boundary without simplification')
plt.show()
plt.close()
````

````python
sns.displot([p.shape[0] for p in simplified_nuclei_polys])
plt.title('Distribution of number of vertices per nuclei boundary simplified')
plt.show()
plt.close()
````
