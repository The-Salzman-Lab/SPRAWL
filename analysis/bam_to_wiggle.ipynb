# Split SS2 main BAM by annotation

```python
import pandas as pd
import numpy as np
import pysam
import glob
import sys
import os


#Sam flag meanings from https://broadinstitute.github.io/picard/explain-flags.html
bit_meanings = {
    'read paired':                               0b000000000001,
    'read mapped in proper pair':                0b000000000010,
    'read unmapped':                             0b000000000100,
    'mate unmapped':                             0b000000001000,
    'read reverse strand':                       0b000000010000,
    'mate reverse strand':                       0b000000100000,
    'first in pair':                             0b000001000000,
    'second in pair':                            0b000010000000,
    'not primary alignment':                     0b000100000000,
    'read fails platform/vendor quality checks': 0b001000000000,
    'read is PCR or optical duplicate':          0b010000000000,
    'supplementary alignment':                   0b100000000000,
}

#If any of these bits are set, reject the read
#can test if a flag is rejectable if flag&reject_flags != 0
reject_flags = (
    bit_meanings['read unmapped'] |
    bit_meanings['mate unmapped'] |
    bit_meanings['second in pair'] |
    bit_meanings['not primary alignment'] |
    bit_meanings['read fails platform/vendor quality checks'] |
    bit_meanings['supplementary alignment'] |
    bit_meanings['read is PCR or optical duplicate']
)

def pass_filter(read):
    cigar = read.cigar
    mq = read.mapping_quality
    read_length = read.query_length
    rejectable_flag = read.flag & reject_flags
    if cigar == [(0, read_length)] and mq == 255 and not rejectable_flag:
        return True


bam_path = '/scratch/groups/horence/rob/data/MERFISH_scRNAseq/ss2_post_picard_bam_tag_filtering/SS2_merged_filtered.bam'
meta_path = '/oak/stanford/groups/horence/rob/readzs_fork/MOp_SS2_metadata.tsv'

meta_df = pd.read_csv(meta_path,sep='\t')
meta_df['subclass_label'] = meta_df['subclass_label'].str.replace('/','')
meta_df['subclass_label'] = meta_df['subclass_label'].str.replace(' ','_')
meta_df['cell_id'] = meta_df['cell_id'].str.replace('SS2_','')

bc_to_ann = dict(meta_df[['cell_id','subclass_label']].values)


#Get file handle to bam to split
in_bam = pysam.AlignmentFile(bam_path, 'rb')


#Prepare output bams for each possible cell-type
out_stem = '/scratch/groups/horence/rob/data/MERFISH_scRNAseq/ss2_post_picard_bam_tag_filtering/split_by_celltype'
out_bams = {}
out_paths = []
for ann in meta_df['subclass_label'].unique():
    out_path = os.path.join(out_stem,ann+'.bam')
    out_bam = pysam.AlignmentFile(out_path,'wb',template=in_bam)
    out_bams[ann] = out_bam
    out_paths.append(out_path)

#Iterate through the input bam
for i,r in enumerate(in_bam.fetch()):
    try:
        bc = r.get_tag('CB')
    except KeyError:
        continue

    ann = bc_to_ann.get(bc)
    if ann and pass_filter(r):
        out_bams[ann].write(r)


#Close the output bams
for ann,bam_f in out_bams.items():
    bam_f.close()


#Index the bams
for out_path in out_paths:
    pysam.index(out_path)
```
