```python
#Average number of spots/genes per cell per sample
#this is very slow, so just iterate through max of 5000 cells to get an estimate per sample
importlib.reload(sprawl)

data = {
    'sample':[],
    'num_cells':[],
    'num_genes':[],
    'avg_genes_per_cell':[],
    'avg_spots_per_cell':[],
    'avg_zslice_per_cell':[],
}

for hdf5_path in sorted(glob.glob('../../inputs/hdf5s/*.hdf5')):
    sample = sprawl.HDF5(hdf5_path)
    
    num_cells = 0
    num_genes = 0
    num_spots = 0
    num_zs = 0
    
    start = time.time()
    for i,cell in enumerate(sample.iter_cells()):
        if i >= 5000:
            break
            
        num_cells += 1
        num_genes += len(cell.genes)
        num_spots += sum(cell.n_per_z.values())
        num_zs += len(cell.zslices)
        
    data['sample'].append(os.path.basename(hdf5_path))
    data['num_cells'].append(num_cells)
    data['num_genes'].append(sample.num_genes)
    data['avg_genes_per_cell'].append(num_genes/num_cells)
    data['avg_spots_per_cell'].append(num_spots/num_cells)
    data['avg_zslice_per_cell'].append(num_zs/num_cells)
    
    print(hdf5_path,time.time()-start)
    
    
df = pd.DataFrame(data)
df
```
