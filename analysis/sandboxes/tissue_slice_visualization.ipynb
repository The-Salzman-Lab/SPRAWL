```python
plotting.plot_tissue_level(
    cells,
    color_by_ontology=True,
    region=(0, 0, 50, 50), #limit to just one of the slices
)
plt.savefig('mop1_tissue.svg')
plt.show()
plt.close()
```

```python
for i,cell in enumerate(cells):
    if cell.n < 200:
        continue
        
    fig,ax = plotting.plot_cell_3D(cell, rainbow=True)
    ax.legend([])
    ax.view_init(elev=90, azim=0)
    plt.savefig(f'individual_cells/cell_{i}.svg')
    plt.show()
    plt.close()
    
    if i > 10:
        break
```
