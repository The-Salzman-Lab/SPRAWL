```python
ks = [0.04,0.2,1,5,50]
num_its = 100
med_ranks = {}

for k in ks:
    
    gene_ks={
        'C1ql3':k, #biased towards smaller ranks
    }
    
    med_ranks[k] = []

    for it in range(num_its):

        cell = simulate.lin_permute_gene_labels(
            cell,
            metric='peripheral',
            gene_ks=gene_ks,
        )
        med_ranks[k].append(cell.gene_med_ranks['C1ql3'])

med_ranks
```

```python
ks = [-1,-0.5,0,0.5,1]
num_its = 100
med_ranks = {}

for k in ks:
    
    gene_ks={
        'C1ql3':k, #biased towards smaller ranks
    }
    
    med_ranks[k] = []

    for it in range(num_its):

        cell = simulate.exp_permute_gene_labels(
            cell,
            metric='peripheral',
            gene_ks=gene_ks,
        )
        med_ranks[k].append(cell.gene_med_ranks['C1ql3'])

med_ranks
```

```python
plot_df = pd.DataFrame(med_ranks)
plot_df = plot_df.melt(var_name='k',value_name='med_rank')
plot_df['k'] = plot_df['k'].astype(str)

sns.displot(
    x = 'med_rank',
    hue = 'k',
    kind = 'kde',
    data = plot_df,
)
plt.title('Increasing k for linear permutations results in smaller median gene ranks')
plt.show()
plt.close()
```
